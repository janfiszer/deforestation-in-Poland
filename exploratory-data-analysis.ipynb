```python
# essentials
import numpy as np
import os

import tensorflow as tf

from utils import config, visualization
```

```python
masks_dir = os.path.join(config.DATASET_DIR, "masks")
images_dir = os.path.join(config.DATASET_DIR, "images")

image_files = os.listdir(images_dir)
mask_files = os.listdir(masks_dir)
```

```python
dataset_size = len(mask_files)
dataset_size
```

```python
index = np.random.randint(dataset_size)

mask = tf.io.read_file(os.path.join(masks_dir, mask_files[index]))
mask = tf.image.decode_jpeg(mask, channels=3)

image = tf.io.read_file(os.path.join(images_dir, image_files[index]))
image = tf.image.decode_jpeg(image, channels=3)
image = tf.image.convert_image_dtype(image, dtype=tf.float32)

visualization.display([image, mask])
```
